```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'C:\\Users\\CHETHAN\\Downloads\\healthcare_data.csv'  # Replace with your downloaded CSV file path
data = pd.read_csv(file_path)

print(data.head())  # Preview the data
```

```python
# Count the number of patients and doctors
entity_counts = data['Entity'].value_counts()
print(entity_counts)

# Analyze patients by disease
patients = data[data['Entity'] == 'Patient']
disease_counts = patients['Disease'].value_counts()
print(disease_counts)
```

```python
# Pie chart for entity distribution
entity_counts.plot(kind='pie', autopct='%1.1f%%', title='Distribution of Patients and Doctors')
plt.show()

# Bar chart for patients by disease
disease_counts.plot(kind='bar', title='Patients by Disease', color='skyblue')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.show()
```

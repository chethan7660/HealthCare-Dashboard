```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Load the CSV file into a DataFrame
# Replace 'path_to_your_file.csv' with the actual path to your CSV file
df = pd.read_csv('C:\\Users\\CHETHAN\\Downloads\\hospital-data.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Data Cleaning: Check for missing values
print(df.isnull().sum())
```

```python
# Visualize the distribution of hospital ownership types
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Hospital Ownership', order=df['Hospital Ownership'].value_counts().index, palette='viridis')
plt.title('Distribution of Hospital Ownership Types')
plt.xticks(rotation=45)
plt.ylabel('Number of Hospitals')
plt.xlabel('Ownership Type')
plt.show()
```

```python
# Visualize the number of hospitals by state
plt.figure(figsize=(12, 6))
state_counts = df['State'].value_counts()
sns.barplot(x=state_counts.index, y=state_counts.values, palette='magma')
plt.title('Number of Hospitals by State')
plt.xticks(rotation=90)
plt.ylabel('Number of Hospitals')
plt.xlabel('State')
plt.show()
```

```python
# Visualize the relationship between hospital type and emergency services
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Hospital Type', hue='Emergency Services', palette='coolwarm')
plt.title('Hospital Type vs Emergency Services')
plt.xticks(rotation=45)
plt.ylabel('Number of Hospitals')
plt.xlabel('Hospital Type')
plt.legend(title='Emergency Services', loc='upper right')
plt.show()
```

```python
# Visualize the top 10 hospitals by phone number (as a proxy for size)
top_hospitals = df.nlargest(10, 'Phone Number')  # Assuming Phone Number is numeric for this example
plt.figure(figsize=(12, 6))
sns.barplot(data=top_hospitals, x='Hospital Name', y='Phone Number', palette='crest')
plt.title('Top 10 Hospitals by Phone Number')
plt.xticks(rotation=45)
plt.ylabel('Phone Number')
plt.xlabel('Hospital Name')
plt.show()
```
